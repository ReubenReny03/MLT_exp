```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fcmeans import FCM


# 1. Load the dataset
data = pd.read_csv('iris.csv')

data_features = data.drop('variety', axis=1)

# 4. Plot the data points using scatter plots
plt.scatter(data_features.iloc[:, 0], data_features.iloc[:, 1], alpha=0.5)
plt.show()

n = 3

fcm = FCM(n_clusters=n)
fcm.fit(data_features.values)
# Outputs
fcm_centers = fcm.centers
fcm_labels = fcm.u.argmax(axis=1)

for i in range(n):
    plt.scatter(data_features.values[fcm_labels == i, 0], 
                data_features.values[fcm_labels == i, 1])
plt.scatter(fcm_centers[:, 0], fcm_centers[:, 1], s=100, c='red',)

plt.show()

# Performance metrics
partition_coefficient = fcm.partition_entropy_coefficient
print("Partition Coefficient:", partition_coefficient)
# classification entropy is not working
```
