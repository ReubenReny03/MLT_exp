```python
import pandas as pd
from sklearn.preprocessing import Normalizer
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df
normala = Normalizer()

test = normala.fit_transform(df.iloc[:,:-1])
X = pd.DataFrame(test,columns=["sepal.length","sepal.width","petal.length","petal.width"])
```

```python
X
```

```python
linked = linkage(X)
dendrogram(linked)
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

def apply_clustering(X, n_clusters=3, linkage_method='ward', metric='euclidean'):
    cluster = AgglomerativeClustering(n_clusters=n_clusters, metric=metric, linkage=linkage_method)
    return cluster.fit_predict(X)

euclidian_labels_one = apply_clustering(X[['petal.length']], linkage_method='complete', metric='euclidean')
labels_average_manhattan = apply_clustering(X[['petal.length']], linkage_method='average', metric='manhattan')
labels_single_cosine = apply_clustering(X[['petal.length']], linkage_method='single', metric='cosine')

```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

def apply_clustering(X, n_clusters=3, linkage_method='ward', metric='euclidean'):
    cluster = AgglomerativeClustering(n_clusters=n_clusters, metric=metric, linkage=linkage_method)
    return cluster.fit_predict(X)

euclidian_labels_complete = apply_clustering(X, linkage_method='complete', metric='euclidean')
labels_average_manhattan = apply_clustering(X, linkage_method='average', metric='manhattan')
labels_single_cosine = apply_clustering(X, linkage_method='single', metric='cosine')

```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=euclidian_labels_complete, cmap='rainbow')
plt.title('Cluster Visualization with Full Data')
plt.show()
```
